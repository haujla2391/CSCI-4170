#Web-based Data Visualization

```python
import pandas as pd
import numpy as np
# Must mount in drive or attach file to session
df = pd.read_csv("/content/sample_data/movies.csv")

print(df.head())
print(df.info())
```

```python
# Plot y=feature 10 over x=feature 6
# Plot y=feature 8 over x=feature 6
import matplotlib.pyplot as plt
good = df[df["IsGoodRating"] == 1]
bad  = df[df["IsGoodRating"] == 0]

plt.scatter(good["imdbRating"], good["WinsNoms"], color='blue', label='Good Rating (1)', s=30, marker='+')
plt.scatter(bad["imdbRating"], bad["WinsNoms"], color='red', label='Bad Rating (0)', s=30)

plt.title("WinsNoms vs imdbRating")
plt.xlabel("imdbRating")
plt.ylabel("WinsNoms")
plt.legend()
plt.show()
```

```python
good = df[df["IsGoodRating"] == 1]
bad  = df[df["IsGoodRating"] == 0]

plt.scatter(good["imdbRating"], good["Budget"], color='blue', label='Good Rating (1)', s=30, marker='+')
plt.scatter(bad["imdbRating"], bad["Budget"], color='red', label='Bad Rating (0)', s=30)

plt.title("WinsNoms vs Budget")
plt.xlabel("imdbRating")
plt.ylabel("Budget")
plt.legend()
plt.show()
```

```python
# Plot y=feature 10 over x=feature 6
# Plot y=feature 8 over x=feature 6
import matplotlib.pyplot as plt
good = df[df["IsGoodRating"] == 1]
bad  = df[df["IsGoodRating"] == 0]
plt.scatter(good["imdbRating"], good["WinsNoms"], color='blue', label='Good Rating (1)', s=good['WinsNoms'], marker='+')
plt.scatter(bad["imdbRating"], bad["WinsNoms"], color='red', label='Bad Rating (0)', s=bad['WinsNoms'])
plt.title("WinsNoms vs imdbRating")
plt.xlabel("imdbRating")
plt.ylabel("WinsNoms")
plt.legend()
plt.show()
```

```python
# Plot y=feature 10 over x=feature 6
# Plot y=feature 8 over x=feature 6
import matplotlib.pyplot as plt
good = df[df["IsGoodRating"] == 1]
bad  = df[df["IsGoodRating"] == 0]
plt.scatter(good["imdbRating"], np.sqrt(good["imdbVotes"]), color='blue', label='Good Rating (1)', s=good['WinsNoms'], marker='+')
plt.scatter(bad["imdbRating"], np.sqrt(bad["imdbVotes"]), color='red', label='Good Rating (1)', s=bad['WinsNoms'])
plt.title("Votes vs imdbRating")
plt.xlabel("imdbRating")
plt.ylabel("Sqrt Votes")
plt.legend()
plt.show()
```

```python
# Plot y=feature 10 over x=feature 6
# Plot y=feature 8 over x=feature 6
import matplotlib.pyplot as plt
good = df[df["IsGoodRating"] == 1]
bad  = df[df["IsGoodRating"] == 0]
plt.scatter(good["imdbRating"], np.log(good["imdbVotes"]), color='blue', label='Good Rating (1)', s=good['WinsNoms'], marker='+')
plt.scatter(bad["imdbRating"], np.log(bad["imdbVotes"]), color='red', label='Good Rating (1)', s=bad['WinsNoms'])
plt.title("Votes vs imdbRating")
plt.xlabel("imdbRating")
plt.ylabel("Log Votes")
plt.legend()
plt.show()
```
